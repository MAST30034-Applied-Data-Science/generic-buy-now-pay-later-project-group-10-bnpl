Downloading Data for Inflation 

```python
#Downloading data for Inflation 


output_directory = '/Users/ahirve/Desktop/ads_bnpl/data/'

inflation_dir = 'inflation'

if not os.path.exists(output_directory + merch_fees_dir):
    os.makedirs(output_directory + merch_fees_dir)

inflation_url = 'https://www.rba.gov.au/statistics/tables/csv/g1-data.csv?v=2022-09-16-09-09-45'

output_directory = f"{output_directory}/{'inflation'}.csv"
urlretrieve(inflation_url, output_directory)
```

```python
inflation_pd = pd.read_csv('/Users/ahirve/Desktop/ads_bnpl/data/inflation.csv',encoding='cp1252')[1:]
inflation_pd.head(20)
```

```python
#Extracting required columns and creating dataframe 

dates = inflation_pd["G1 CONSUMER PRICE INFLATION"][14:]
percentage_inflation = inflation_pd["Unnamed: 2"][14:]
```

```python
dates.to_list
percentage_inflation.to_list
```

```python
inflation_cleaned_pd = pd.DataFrame()
inflation_cleaned_pd["dates"] = dates
inflation_cleaned_pd["percentage_inflation"] = percentage_inflation
```

```python
inflation_cleaned_pd.head(5)
```

```python
dates_pd = inflation_cleaned_pd['dates'].str.split('-', expand = True)
dates_pd.head(5)
```

```python
inflation_cleaned_pd['month'] = dates_pd[0]
inflation_cleaned_pd['year'] = dates_pd[1]

```

```python
#Converting year to integer 
year_col = dates_pd[1]
year_col.to_list()
year_col = [int(item) for item in year_col]

percentage_inflation_col = inflation_cleaned_pd['percentage_inflation']
percentage_inflation_col = [float(item) for item in percentage_inflation_col]
```

```python
inflation_cleaned_pd['year'] = year_col
inflation_cleaned_pd['month'] = dates_pd[0]
inflation_cleaned_pd['percentage_inflation'] = percentage_inflation_col
```

```python
inflation_cleaned_pd.head(5)
```

```python
#further cleaning: removing all data taken before the year 2020, and removing NaN values for data not available yet: 

inflation_cleaned_pd = inflation_cleaned_pd.loc[inflation_cleaned_pd['year']>=2020]

inflation_cleaned_pd = inflation_cleaned_pd[inflation_cleaned_pd['percentage_inflation'].notna()]
```

```python
inflation_cleaned_pd.head(15)
```

```python
#Plotting line graph to visualize trends in inflation: 
plt.plot(inflation_cleaned_pd['year'],inflation_cleaned_pd['percentage_inflation'])
plt.title('Trend for Inflation, 2020 to 2022')
plt.xlabel('Dates')
plt.ylabel('Percentage Inflation')
plt.show()
plt.savefig('/Users/ahirve/Desktop/ads_bnpl/plots/inflation_trends.png')
```

```python
#Aggregating inflation percentages for each year 

inflation_agg = inflation_cleaned_pd.groupby('year')['percentage_inflation'].aggregate('mean')
inflation_agg.head(5)
```
