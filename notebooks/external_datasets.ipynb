Downloading external datasets 

```python
import sys 
import os 
from urllib.request import urlretrieve
import pandas as pd 
```

Downloading Exchange rates data 

```python
#Creating output directories
output_directory = '/Users/ahirve/Desktop/ads_bnpl/data/'

exc_rate_dir = 'exchange_rates'

if not os.path.exists(output_directory + exc_rate_dir):
    os.makedirs(output_directory + exc_rate_dir)

exc_rate_url = 'https://www.rba.gov.au/statistics/tables/csv/f11.1-data.csv'

output_directory = f"{output_directory}/{exc_rate_dir}.csv"
urlretrieve(exc_rate_url, output_directory)

```
