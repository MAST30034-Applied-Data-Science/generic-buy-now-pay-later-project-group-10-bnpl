Downloading external datasets 

```python
import sys 
import os 
from urllib.request import urlretrieve
import pandas as pd 
```

Downloading data for Average Merchant Fees for Debit, Credit, and Exchange Cards

```python

output_directory = '/Users/ahirve/Desktop/ads_bnpl/data/'

merch_fees_dir = 'merch_fees'

if not os.path.exists(output_directory + merch_fees_dir):
    os.makedirs(output_directory + merch_fees_dir)

merch_fees_url = 'https://www.rba.gov.au/statistics/tables/xls/c03hist.xlsx?v=2022-09-09-07-37-45'

output_directory = f"{output_directory}/{'merch_fees'}.xls"
urlretrieve(merch_fees_url, output_directory)

```
